# Q1. 寫一個函式，抓取 Google Finance 的股價歷史資料

```python
!pip install requests
import requests 
import pandas as pd
import io
```

```python
def goog(stock,date):
        url = 'http://finance.google.com/finance/historical'
        params = {}
        params['q'] = stock
        params['startdate'] = date
        params['output'] = 'csv'
        
        
        price = requests.get(url,params=params)
        Data = pd.read_csv(io.StringIO(price.text))
        return Data
```

```python
df = goog('goog','2016-1-1')
print(df.tail())
```
