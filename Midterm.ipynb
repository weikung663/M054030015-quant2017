# Q1. 畫出 Diamond Shape

```python
def diamond(n):
 print(" "*(n-1)+"Λ")
 for i in range(0,n-1):
        print(" "*(n-i-2)+"╱"+" "*(2*i+1)+"╲")
 for j in reversed(range(0,n-1)):
        print(" "*(n-j-2)+"╲"+" "*(2*j+1)+"╱")
 print(" "*(n-1)+"V")
```

```python
diamond(8)
```

```python
#待修正
def RealDiamond(n):
    print(" "*n+"_"*n)
    for i in range(0,n-2):
        print(" "*(n-i-1)+"╱"+" "*(n+2*i)+"╲")
    for j in reversed(range(0,n)):
        print(" "*(n-j+1)+"╲"+" "*((n-4)+2*j)+"╱")
```

```python
RealDiamond(6)
```

# Q2. 刪除重複性資料

```python
#從 symbols01.txt 跟 symbols02.txt 讀入兩組股票代號，合併兩組代號
file = open('symbols01.txt', 'r')
f1 = file.readlines()
file.close()
file = open('symbols02.txt', 'r')
f2 = file.readlines()
f3 = f1+f2
```

```python
#輸出一份無重複的股票代號到 symbols03.txt
File = list(set(f3))
File.sort(key=f3.index)
file = open('symbols03.txt', 'w')
File = ''.join(File)
file.write(File)
file.close()
```

# Q4. 畫出技術曲線。

```python
import pandas_datareader.data as web
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import talib
from datetime import datetime
df = web.DataReader('TSLA', 'yahoo', datetime(2015, 1, 1))
df['Adj Close'].fillna(df['Close'], inplace=True)
df = df.dropna()
df.head()
```

```python
v = np.log10(np.abs(df['Volume'])).values
```

```python
df['obv']= talib.OBV(df["Close"].values,v)
df['rsi']= talib.RSI(df["Close"].values, timeperiod=7)
df.tail()
```

```python
%matplotlib inline
df[['rsi','obv']].plot(legend='best', figsize = (15, 5))
```

```python
df['20d'] = pd.Series.rolling(df['obv'], window=20).mean()
df['60d'] = pd.Series.rolling(df['obv'], window=60).mean()
df[['Close','obv', '20d', '40d']].plot(legend='best', figsize=(15,5))
```

```python
df['20d_r'] = pd.Series.rolling(df['rsi'], window=20).mean()
df['60d_r'] = pd.Series.rolling(df['rsi'], window=60).mean()
df[['Close','rsi', '20d_r', '60d_r']].plot(legend='best', figsize=(15,5))
```
